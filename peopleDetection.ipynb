```python
import numpy as np
import cv2
import winsound


cv2.startWindowThread()
cap = cv2.VideoCapture(0)

while True:

    # reading the frame
    ret, frame = cap.read()

    # Displaying the video? frame
    cv2.imshow('frame',frame)

    # Exit condition(press 'q')
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    # reading back-to-back frames(im/ages) from video
    ret, frame1 = cap.read()
    ret, frame2 = cap.read()

    # Difference between frames
    diff = cv2.absdiff(frame1, frame2)

    # Converting image to a better form -'threshDiff'.
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    _, threshDiff = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)

    # Find contours of moving object
    contours, hirarchy = cv2.findContours(threshDiff, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # draw
    cv2.drawContours(frame, contours, -1, (0,255,255), 3)
    
    #count 
    people = 0
    people = sum(1 for i in contours if cv2.contourArea(i)>1000) # 1000 is my definition of human size.
        
    if people > 0:
        duration = 1000  # milliseconds
        freq = 740  # Hz
        winsound.Beep(freq, duration)
        size = sum(cv2.contourArea(i) for i in contours if cv2.contourArea(i)>1000)
    
        
    # Showing the output Image
    cv2.putText(frame, text = 'Num Of People: ' +str(people)+'size: '+str(size),fontFace =cv2.FONT_HERSHEY_SIMPLEX,
                org = (40,450), fontScale=1.2, color=(255,255,255),thickness=2)
    cv2.imshow("DetectPeople", frame)
    
    summ = 0

# Release cap resource
cap.release()

# Destroy all windows
cv2.destroyAllWindows()
```
