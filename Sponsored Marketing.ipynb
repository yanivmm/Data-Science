```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
sns.set(font_scale=1.2)
palettes = ['husl',"Set2",'pastel','autumn','gnuplot2_r','gnuplot2']
hebrewChannelNames =['']
```

```python
path = r'C:\Users\97250\Desktop\KLogic\‏‏DRIVE-BY - עותק.XLSX'
allData = pd.read_excel(path)
```

```python
allChannels = allData[1:13]

searchBrand = allData[16:28]

searchCompeting = allData[31:43]

searchGenericWords = allData[46:58]

Facebook = allData[61:73]
```

```python
for data in  [allChannels,searchBrand,searchCompeting,searchGenericWords,Facebook]:
    data.columns = "DATE	COST	IMPRESSIONS	CLICKS	CTR	CPC	CONVERSIONS	COST PER CONVERSION	CONVERTION RATE".lower().split("	")
    data.set_index('date',inplace=True)
```

```python
data = pd.concat([allChannels,searchBrand,searchCompeting,searchGenericWords,Facebook],axis=1,keys='allChannels,searchBrand,searchCompeting,searchGenericWords,Facebook'.split(','))
```

```python
data.columns.names = ['Channel','Information']
```

```python
def show(data,title = ''):
    data.plot(figsize=(10,6),fontsize =12,lw=3)
    plt.title('\n'+ title+' between the 1-12 of January \n',fontsize=30)
    plt.ylabel(title +'\n', fontsize=20)
    plt.xlabel('Date',fontsize=20)
    plt.legend(loc='upper right',prop={'size':16})  
```

```python
show(CTR,'CTR')
```

```python
show(CPC,'CPC')
```

```python
show(data['allChannels']['impressions'],'Impression of all channels')
```

```python
def showBar(data,title= 'CPC'):
    
    data.plot(figsize=(8,4),fontsize =12, kind = 'bar', color = ['black','orange','red','y','b'])
    plt.title('\n '+ title +' through time \n',fontsize=20)
    plt.ylabel(title+'\n', fontsize=18)
    plt.xlabel('')
```

```python
CPCmean = CPC.mean()
```

```python
showBar(CPCmean)
```

```python
showBar(costPerConversion.mean(), title = 'Cost Per Conversion')
```

```python
def showBar2(data,col ='Cost Per Conversion'):
    data = data.reset_index()
    data.columns = ['channel',col]
    plt.figure(figsize=(11,4))
    sns.barplot(x=data.columns[0], y=data.columns[1], data=data, palette=palettes[0])
    plt.title('\n' +data.columns[1]+ ' \n',fontsize=25)
    plt.ylabel('')
    plt.xlabel('')   
```

```python
showBar2(costPerConversion.mean())
```

```python
showBar2(conversionRate.mean(),col='Convertion Rate')
```
