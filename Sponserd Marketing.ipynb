```python
def show(data,title = ''):
    data.plot(figsize=(10,6),fontsize =12,lw=3)
    plt.title('\n'+ title+' timeline between the 1-12 of January \n',fontsize=30)
    plt.ylabel(title +'\n', fontsize=20)
    plt.xlabel('Date',fontsize=20)
    plt.legend(loc='upper right',prop={'size':16})  
```

```python
def showBar2(data,col ='Cost Per Conversion'):
    data = data.reset_index()
    data.columns = ['channel',col]
    plt.figure(figsize=(9,4))
    sns.barplot(x=data.columns[0], y=data.columns[1], data=data, palette=palettes[4])
    plt.title('\n' +data.columns[1]+ ' \n',fontsize=25)
    plt.ylabel('')
    plt.xlabel('')   
```

```python
show(CTR,'CTR')
```

```python
def convertData(data):

    def convertor(x):

        def remover(x):
            x = x.replace('k','').replace('m','').replace(',','')
            return float(x)

        x = str(x)

        if 'k' in x:
            return int(remover(x)*1000)

        if 'm' in x:
            return int(remover(x)*1000000)

        else:
            return float(remover(x))

    data['Facebook']             = data['Facebook'].apply(convertor)
    data['searchGenericWords']   = data['searchGenericWords'].apply(convertor)
    data['searchCompeting']      = data['searchCompeting'].apply(convertor)
    data['searchBrand']          = data['searchBrand'].apply(convertor)
    data['allChannels']          = data['allChannels'].apply(convertor)
```

```python
convertData(cost)
convertData(CPC)
```
